## Concatenate , drop and fill in missing values

```python
data= pd.concat([Train,Test],sort=False)
data.info()
```

```python
data['Arrival at Destination - Day of Month'] =data['Arrival at Destination - Day of Month'].fillna(data['Arrival at Destination - Day of Month'].median())
data['Arrival at Destination - Weekday (Mo = 1)'] = data['Arrival at Destination - Weekday (Mo = 1)'].fillna(data['Arrival at Destination - Weekday (Mo = 1)'].median())
data['Arrival at Destination - Time'] = data['Arrival at Destination - Time'].fillna(data['Arrival at Destination - Time'].mode())
data['Temperature'] = data['Temperature'].fillna(data['Temperature'].mean())
data['Time from Pickup to Arrival'] = data['Time from Pickup to Arrival'].fillna(data['Time from Pickup to Arrival'].mean())
```

## Using the random forest. Just trying it out

# creating x and y

```python
X = data_drop2.iloc[:,:-1].values
y = data_drop2.iloc[:,-1].values
print(X)
```

## Splitting test and training set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## training random forest regression set

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

# Predicting y

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Evaluating model

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error 
from math import sqrt

mse = sqrt(mean_squared_error(y_test,y_pred))
print(mse)
```
